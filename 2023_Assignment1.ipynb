## Introduction

The Australian Dream, in its narrowest sense, is to own a house. Imagine in three years, you successfully graduate from the University of Adelaide, and decide to explore other capital cities in Australia. In this scenario it is probably a good idea to have a rough idea about how long will it take you to save enough money for the up-front payment of your dream home in all capital cities. In this coding assignment, you will write a simple calculator for such calculation.

## Specifications
Your program will take the following inputs from the user:
1. The annual salary (for the variable `annual_salary`)
2. The portion of salary to be saved (for the variable `saving_rate`)
3. The price of your dream house (for the variable `home_price`)

The program will then calculate the number of months required to have enough savings (`save_amount`) for the upfront payment of your dream house. Note that the upfront payment required by the bank will include both `down payment` and `stamp duty`.

Below we define a series of variables you will use in the program.

**Cost of a House**
- `home_price`: the cost of your dream home.
- `down_payment_rate`: down payment is the cash you pay upfront when you purchase the property. The remaining amount will go into your home mortgage. For example, assume the down payment rate is `30%`. If you are purchasing a `1,000,000` house, you will pay `300,000` in cash upfront, and borrow a `700,000` mortgage from the bank.
- `stamp_duty_rate`: you need to pay the stamp duty to the state when you purchase a property. For example, assume the stamp duty rate is `4%` and you are purchasing a `1,000,000` house. You will pay `40,000` to the state government.

**Salary and Saving**
- `annual_salary`: Your annual salary. The average salary of a computer programmer is around `80,000`. As a graduate from University of Adelaide, you are highly sought off and your salary will be higher than that!
- `tax_rate`: A portion of your salary will go into tax payment. The calculation of tax is complex and we will just use a fixed tax rate in the program.
- `save_amount`: The amount of money you have saved. 
- `saving_rate`: The percentage of your *after-tax* income that goes into saving. For example, with a tax rate of `30%` and a saving rate of `10%`, the amount you can save per year will be `80,000 * (1-0.3) * 0.1 = 5,600`
- `monthly_return_rate`: This number reflects the monthly profit of your investment. For example, assume at December you have a saved amount of `5000`. You invested your saving smartly and have a rate of return of `0.1%`. It means that your saving will increase by `5000*0.001 = 50` and become `5050`.

```python
# This function accepts a user-supplied argument and outputs a tuple of three values: annual_salary, saving_rate, and house_price.

def set_value():
    global annual_salary, saving_rate, house_price
    annual_salary = float(input("Please enter your annual salary: "))
    saving_rate = float(input("Please indicate how much of your salary you intend on saving: "))
    house_price = float(input("Please enter your house's cost: "))
    return annual_salary, saving_rate, house_price

# This function provides the total upfront payment requested after receiving the cost of the dream home, the down payment rate, and the stamp duty rate as inputs.

def cal_upfront(house_price, down_payment_rate, stamp_duty_rate):
    # INSERT YOUR CODE
    down_payment = house_price * down_payment_rate
    stamp_duty = house_price * stamp_duty_rate
    upfront = down_payment + stamp_duty
    return upfront

# This function calculates the number of months needed to save up enough money for the down payment on the dream home using the annual_salary, saving_rate, and home_price as inputs.

def cal_month_to_save_upfront(up_front):
    # HINT: Calculate these numbers first, then update your saving amount monthly in a loop
    months = 0
    down_payment_rate = 0.3
    stamp_duty_rate = 0.04
    tax_rate = 0.3
    monthly_return_rate = 0.01
    save_amount = 0
    up_front = cal_upfront(house_price, down_payment_rate, stamp_duty_rate)
    
    while save_amount < up_front:
        monthly_salary = annual_salary / 12
        monthly_tax = monthly_salary * tax_rate
        monthly_income = monthly_salary - monthly_tax
        monthly_saving = monthly_income * saving_rate
        save_amount += monthly_saving + save_amount * monthly_return_rate
        months += 1
    return months

# This code receives the user's inputs and then sends the values to the function to find the required number of months.

annual_salary, saving_rate, house_price = set_value()
up_front = cal_upfront(house_price, 0.3, 0.04)
m = cal_month_to_save_upfront(up_front)
print(f'Upfront is {up_front} and it will take {m} months of saving')



# EXAMPLE 1
test_1()
up_front = cal_upfront(house_price, down_payment_rate, stamp_duty_rate)
m = cal_month_to_save_upfront(up_front)
print(f'Upfront is {up_front} and it will take {m} months of saving')
# Upfront is 230000.0 and it will take 99 months of saving

# EXAMPLE 2
test_2()
up_front = cal_upfront(house_price, down_payment_rate, stamp_duty_rate)
m = cal_month_to_save_upfront(up_front)
print(f'Upfront is {up_front} and it will takes {m} months of saving')
# Upfront is 138000.0 and it will take 107 months of saving

# YOUR TEST
set_value()
up_front = cal_upfront(house_price, down_payment_rate, stamp_duty_rate)
m = cal_month_to_save_upfront(up_front)
print(f'Upfront is {up_front} and it will takes {m} months of saving')
```
